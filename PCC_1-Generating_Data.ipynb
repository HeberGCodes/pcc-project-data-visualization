<H3>Changing the label type and line thickness

```python
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()
ax.plot(squares, linewidth=3) # New line

# Set chart title and label axes.
ax.set_title("Square Number", fontsize=24) # New line
ax.set_xlabel("Value", fontsize=14) # New line
ax.set_ylabel("Square of Value", fontsize=14) # New line

# Set size of tick labels.
ax.tick_params(labelsize=14) # New line

plt.show()
```

<H3>Calculating data automatically

```python
import matplotlib.pyplot as plt

x_values = range(1, 1001) # New line
y_values = [x**2 for x in x_values] # New line

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10) # New line

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=24)

# Set size of tick labels.
ax.tick_params(labelsize=14)

# Set the range for each axis.
ax.axis([0, 1100, 0, 1_100_000]) # New line

plt.show()
```

<H3>Customizing tick labels and defining custom colors

```python
import matplotlib.pyplot as plt

x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10) # New line

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=24)

# Set size of tick labels.
ax.tick_params(labelsize=14)

# Set the range for each axis.
ax.axis([0, 1100, 0, 1_100_000])
ax.ticklabel_format(style='plain') # New line.

plt.show()
```

<H5>Cubes: A number raised to the third power is a cube. Plot the first five cubic numbers, and then plot the first 5,000 cubic numbers.

```python
import matplotlib.pyplot as plt

x_values = [1, 2, 3, 4, 5]
y_values = [x**3 for x in x_values]

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=50)

ax.set_title("Cubes", fontsize=28)
ax.set_xlabel("Values", fontsize=14)
ax.set_ylabel("Cubes of Values", fontsize=14)

ax.tick_params(labelsize=14)
ax.ticklabel_format(style='plain')

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(1, 5001)
y_values = [x**3 for x in x_values]

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10)

ax.set_title("Cubes", fontsize=28)
ax.set_xlabel("Values", fontsize=14)
ax.set_ylabel("Cube of Values", fontsize=14)

ax.tick_params(labelsize=14)
ax.axis([1, 5500, 1, 135_000_000_000])
ax.ticklabel_format(style='plain')

plt.show()
```

```python
import matplotlib.pyplot as plt

x_values = range(1, 5001)
y_values = [x**3 for x in x_values]

plt.style.use('seaborn-v0_8')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10)

ax.set_title("Colored Cubes", fontsize=28)
ax.set_xlabel("Values", fontsize=14)
ax.set_ylabel("Cube of Values", fontsize=14)

ax.tick_params(labelsize=14)
ax.axis([1, 5500, 1, 135_000_000_000])
ax.ticklabel_format(style='plain')

plt.show()
```
