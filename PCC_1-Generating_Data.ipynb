### Changing the label type and line thickness

```python
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()
ax.plot(squares, linewidth=3) # New line

# Set chart title and label axes.
ax.set_title("Square Number", fontsize=24) # New line
ax.set_xlabel("Value", fontsize=14) # New line
ax.set_ylabel("Square of Value", fontsize=14) # New line

# Set size of tick labels.
ax.tick_params(labelsize=14) # New line

plt.show()
```

# Random Walks

### Creating the RandomWalk class and choosing directions

```python
from random import choice

class RandomWalk:
    """A class to generate random walks."""
    
    def __init__(self, num_points=5000):
        """Initialize attributes of a walk."""
        self.num_points = num_points

        # All walks start at (0,0).
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        """Calculate all the points in the walk."""

        # Keep taking steps until the walk reaches the desired length.
        while len(self.x_values) < self.num_points:

            x_step = self.get_step()
            y_step = self.get_step()

            # Reject moves that go nowhere.
            if x_step == 0 and y_step == 0:
                continue

            # Calculate the new position.
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step

            self.x_values.append(x)
            self.y_values.append(y)

    def get_step(self):
        """Determines the direction and distance for each step."""

        # Decide which direction to go, and how far to go.
        direction = choice([1, -1])
        distance = choice([0, 1, 2, 3, 4])
        step = direction * distance

        return step
```

### Plotting the random walk

```python
import matplotlib.pyplot as plt

# Keep making new walks, as long as the program is active.
while True:

    # Make a random walk.
    rw = RandomWalk()
    rw.fill_walk()

    # Plot the points in the walk.
    plt.style.use('classic')
    fig, ax = plt.subplots()
    ax.scatter(rw.x_values, rw.y_values, s=15)
    ax.set_aspect('equal')

    plt.show()

    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

#### Styling the Walk:
* Coloring the points
* Plotting the starting and ending points
* Cleaning up the axes

```python
import matplotlib.pyplot as plt

# Keep making new walks, as long as the program is active.
while True:

    # Make a random walk.
    rw = RandomWalk()
    rw.fill_walk()

    # Plot the points in the walk.
    plt.style.use('classic')
    fig, ax = plt.subplots()
    point_numbers = range(rw.num_points) # Coloring the points (this and the next line)
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolors='none', s=15) # Coloring the points
    ax.set_aspect('equal')

    # Emphasize the first and last points.
    ax.scatter(0, 0, c='green', edgecolors='none', s=50)
    ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=50)

    # Remove the axes.
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    plt.show()

    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

#### 
* Adding plot points
* Altering the size to fill the screen

```python
import matplotlib.pyplot as plt

# Keep making new walks, as long as the program is active.
while True:

    # Make a random walk.
    rw = RandomWalk(50_000) # Adding more plot points.
    rw.fill_walk()

    # Plot the points in the walk.
    plt.style.use('classic')
    fig, ax = plt.subplots(figsize=(10, 6), dpi=128) # Altering the size to fill the screen.
    point_numbers = range(rw.num_points)
    ax.scatter(rw.x_values, rw.y_values, c=point_numbers, cmap=plt.cm.Blues, edgecolors='none', s=1) 
    ax.set_aspect('equal')

    # Emphasize the first and last points.
    ax.scatter(0, 0, c='green', edgecolors='none', s=50)
    ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=50)

    # Remove the axes.
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    plt.show()

    keep_running = input("Make another walk? (y/n): ")
    if keep_running == 'n':
        break
```

### Try it yourself

#### Molecular motion:

```python
import matplotlib.pyplot as plt

rw = RandomWalk(5_000)
rw.fill_walk()

plt.style.use('classic')
fig, ax = plt.subplots(figsize=(10, 6), dpi=128)
ax.plot(rw.x_values, rw.y_values, c='blue', linewidth=0.5)
ax.set_aspect('equal')
ax.set_title('Molecular Motion', fontsize=12)

ax.scatter(0, 0, c='green', edgecolors='none', s=66)
ax.scatter(rw.x_values[-1], rw.y_values[-1], c='red', edgecolors='none', s=66)

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

plt.show()
```
